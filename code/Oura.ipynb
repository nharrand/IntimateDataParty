```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the CSV file
df = pd.read_csv("data/oura.csv", parse_dates=['date'])print(df.head(10))
```

```python
print(df.head(10))
```

```python
# Filter columns that contain 'Score'
score_cols = [col for col in df.columns if 'Score' in col]
```

```python
# Compute mean and 95% confidence interval for each score column
means = df[score_cols].mean()
stds = df[score_cols].std()
n = df[score_cols].count()
conf_interval = 1.96 * (stds / np.sqrt(n))
```

```python
# Plot
plt.figure(figsize=(14, 6))
plt.bar(score_cols, means, yerr=conf_interval, capsize=5)
plt.xticks(rotation=90)
plt.ylabel("Average Score")
plt.title("Average Scores with 95% Confidence Intervals")
plt.tight_layout()
plt.show()
```
